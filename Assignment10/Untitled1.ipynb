```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

path = "/media/rifat/STUDY/4-1/LAB/Image_Processing/image/2.jpg"

#img = plt.imread(path)
#gray = cv.cvtColor(img,cv.COLOR_RGB2GRAY)
gray = cv.imread(path,0)

hist = cv.calcHist([gray],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

th,binary = cv.threshold(gray,125,255,cv.THRESH_BINARY_INV)

equ = cv.equalizeHist(gray)


plt.figure(figsize = (10,10))
plt.subplot(1,2,1)
plt.imshow(gray,cmap = 'binary')
plt.title("Original")


plt.subplot(1,2,2)
plt.imshow(equ,cmap='gray')
plt.title("Equalized Image")

plt.show()

hist = cv.calcHist([equ],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()


plt.hist(equ.ravel(),256,(0,256))
plt.show()
```
