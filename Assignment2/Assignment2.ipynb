## Import Libary

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2  as cv
```

## Initialize Value

```python
r = np.random.randint(0,255,size = (10,10))
path = "/media/rifat/STUDY/4-1/LAB/Image_Processing/image/paddy.jpeg"
img = plt.imread(path)
m,n,cn = img.shape

r = cv.cvtColor(img, cv.COLOR_RGB2GRAY)

T1 = 100
T2 = 150
c = 2
p = 5
epsilon = np.finfo(float).eps
print(r)
```

## Originial image with Histogram

```python
fig, ax = plt.subplots(1,2, figsize=(20,6))
plt.subplot(1,2,1)
plt.title("Originial Image")
plt.imshow(r,cmap = 'gray')
d = r.flatten()
plt.subplot(1,2,2)
plt.title("Histrogram of Pixel value")
plt.hist(d,bins = 256)


plt.show()
```

# s = 100, if r >= T1 and r <= T2; otherwise s = 10.

```python
s = []
sz = d.size

for i in range (sz):
    if d[i]>=T1 and d[i]<= T2:   
        s.append(100)
    else:
        s.append(10)          
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,6))
s = np.reshape(s,(m,n))
plt.subplot(1,3,1)
plt.title("Originial Image")
plt.imshow(r,cmap = 'gray')

plt.subplot(1,3,2)
plt.title("Case1 Image")
plt.imshow(s,cmap = 'gray')
s = s.flatten()
plt.subplot(1,3,3)
plt.title("Histrogram of Case1")
plt.hist(s,bins = 25)


plt.show()
```

# s = 100, if r >= T1 and r <= T2; otherwise  s = r.

```python
s = []
for i in range (sz):
    if d[i]>=T1 and d[i]<= T2:       
        s.append(100)
    else:
        s.append(d[i])    
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,6))
s = np.reshape(s,(m,n))
plt.subplot(1,3,1)
plt.title("Originial Image")
plt.imshow(r,cmap = 'gray')

plt.subplot(1,3,2)
plt.title("Case2 Image")
plt.imshow(s,cmap = 'gray')
s = s.flatten()
plt.subplot(1,3,3)
plt.title("Histrogram of Case2")
plt.hist(s)

plt.show()
```

# s = c log(1 + r)  and s = c ( s + epsilon ) ^ p

```python
s = []
s1 = []
for i in range (sz):
    x = c * np.log(1+d[i])
    s.append(x)
    x = c * pow(( d[i] + epsilon ),p)
    s1.append(x)
```

```python
# plt.subplots(2,2, figsize=(8,8))
plt.figure(figsize = (10,10))
plt.subplot(2,2,2)
plt.title("Histrogram of Case3")
plt.hist(s)

s = np.reshape(s,(m,n))
plt.subplot(2,2,1)
plt.title("Case3 Image")
plt.imshow(s,cmap = 'gray')

plt.subplot(2,2,4)
plt.title("Histrogram of Case4")
plt.hist(s1)

s1 = np.reshape(s1,(m,n))
plt.subplot(2,2,3)
plt.title("Case4 Image")
plt.imshow(s1,cmap = 'gray')

plt.show()
```
